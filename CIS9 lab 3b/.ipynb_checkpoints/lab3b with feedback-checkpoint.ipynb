4. Machine Learning: Determine if the data can be separated into groups with similar characteristics.

4a. __Create the X dataset__ and __show the first 5 rows__.

```python
X = data
X.head()
```

```python
### X should come from the cleaned DataFrame of step 2e. It shouldn't be
### the original data from the file          -1/2pt
```

4b. Using the KMeans Clustering model, __determine the best number of clusters__<br>
and __train and test the model__.<br>

_Feel free to create multiple code cells as needed_.

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++',  n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
fig = plt.figure(figsize = (5,5))
plt.plot(np.arange(1,11), wcss)
plt.title("Elbow curve")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

4c. Now that the ML model has created clusters, we want to find the common characteristics of the cluster with the highest number of births.<br>
- Create one or more code cells to __identify the cluster with the highest number of births__.
- Create a Raw NBConvert cell to __discuss which cluster number__ has the highest number of births.

_Note that since there are more than 3 features that the model worked with, it's not possible to plot the clusters in 3D space to visually choose the cluster with the highest number of births_.<br>

```python
for i in range(4):
    print("Cluster", i, "mean number of births:", X[y_kmeans == i].births.mean())
```
